# Task 1: Data Preprocessing

For this experiment we will not be looking for the Description and the missing values in the CustomerId


```python
data_frame.dropna(inplace=True, axis=0)
print('dataFrame.shape: ', data_frame.shape)
data_frame
```

### Remove duplicate records


```python
data_frame = data_frame.drop_duplicates(keep='first')

print('dataFrame.shape: ', data_frame.shape)
data_frame
```

Extract and create new features, such as 'Total Price' = Quantity \* UnitPrice


```python
data_frame['Total Price'] = data_frame['Quantity'] * data_frame['UnitPrice']

mean_total_price = data_frame['Total Price'].mean()
data_frame['Is_High_Value'] = (data_frame['Total Price'] > mean_total_price).astype(int)

data_frame
```

```python
from sklearn.preprocessing import StandardScaler

numerical_columns = ['Quantity', 'UnitPrice', 'Total Price']
scaler = StandardScaler()
data_frame[numerical_columns] = scaler.fit_transform(data_frame[numerical_columns])

print("\nNormalized numerical features:")
data_frame[numerical_columns].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
plt.style.use("seaborn-v0_8-whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Feature Distributions and Relationships', fontsize=16)

# Distribution of normalized numerical features
sns.histplot(data=data_frame['Quantity'], ax=axes[0,0], kde=True)
axes[0,0].set_title('Distribution of Normalized Quantity')
axes[0,0].set_xlabel('Normalized Quantity')

sns.histplot(data=data_frame['UnitPrice'], ax=axes[0,1], kde=True) 
axes[0,1].set_title('Distribution of Normalized Unit Price')
axes[0,1].set_xlabel('Normalized Unit Price')

# Scatter plot of Quantity vs UnitPrice
sns.scatterplot(data=data_frame, x='Quantity', y='UnitPrice', hue='Is_High_Value', 
                alpha=0.5, ax=axes[1,0])
axes[1,0].set_title('Quantity vs Unit Price')

# Box plot of Total Price by Country
sns.boxplot(data=data_frame, y='Total Price', x='Country', ax=axes[1,1])
axes[1,1].set_title('Total Price Distribution by Country')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nSummary Statistics by Country:")
print(data_frame.groupby('Country')[numerical_columns].agg(['mean', 'std', 'count']))

print("\nHigh Value Orders by Country:")
print(data_frame.groupby('Country')['Is_High_Value'].value_counts(normalize=True).unstack())
```
