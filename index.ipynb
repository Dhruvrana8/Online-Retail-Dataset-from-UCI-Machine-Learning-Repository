# Customer Clustering using K-Means


Select relevant features for clustering


```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

# Select features for clustering
clustering_features = ['Quantity', 'UnitPrice', 'Total Price']

# Normalize the selected features
scaler = StandardScaler()
data_frame[clustering_features] = scaler.fit_transform(data_frame[clustering_features])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
data_frame['Cluster'] = kmeans.fit_predict(data_frame[clustering_features])

print(data_frame[['CustomerID', 'Cluster']].head(10))

# Create a figure with 3D plot for cluster visualization
fig = plt.figure(figsize=(15, 10))

# 3D scatter plot showing all three features colored by cluster
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data_frame['Quantity'], 
                    data_frame['UnitPrice'], 
                    data_frame['Total Price'],
                    c=data_frame['Cluster'],
                    cmap='Accent',
                    alpha=0.6)

ax.set_xlabel('Quantity')
ax.set_ylabel('Unit Price')
ax.set_zlabel('Total Price')
ax.set_title('3D Visualization of Customer Clusters')

# Add a color bar
plt.colorbar(scatter, label='Cluster')

# Adjust the viewing angle for better visualization
ax.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()

# Print cluster statistics
print("\nCluster Statistics:")
print(data_frame.groupby('Cluster')[clustering_features].agg(['mean', 'std']).round(2))
```

Normalize the features before applying K-Means


```python
# Select additional features for customer segmentation
clustering_features = ['Quantity', 'UnitPrice', 'Total Price', 'Is_High_Value']

# Normalize the features again with the new selection
scaler = StandardScaler()
data_frame[clustering_features] = scaler.fit_transform(data_frame[clustering_features])

# Apply K-Means clustering with the updated features
kmeans = KMeans(n_clusters=5, random_state=42)
data_frame['Cluster'] = kmeans.fit_predict(data_frame[clustering_features])

# Print cluster statistics for the new features
print("\nUpdated Cluster Statistics:")
print(data_frame.groupby('Cluster')[clustering_features].agg(['mean', 'std']).round(2))

# Create a pair plot to visualize relationships between features
plt.figure(figsize=(15, 12))
for i in range(5):
    cluster_data = data_frame[data_frame['Cluster'] == i]
    plt.subplot(2, 3, i+1)
    plt.scatter(cluster_data['Quantity'], cluster_data['Total Price'], alpha=0.5)
    plt.title(f'Cluster {i}')
    plt.xlabel('Quantity')
    plt.ylabel('Total Price')

plt.tight_layout()
plt.show()

```

Use the Elbow Method to determine the optimal number of clusters


```python
import numpy as np
# Apply the Elbow Method to find optimal number of clusters
inertias = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_frame[clustering_features])
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()

# Calculate the rate of inertia change
inertia_changes = np.diff(inertias)
relative_changes = inertia_changes[:-1] / inertia_changes[1:]

print("\nRelative changes in inertia between consecutive k values:")
for k, change in enumerate(relative_changes, start=1):
    print(f"k={k+1}: {change:.2f}")
```

Apply the K-Means algorithm and interpret the clusters


```python
# Apply K-means with the optimal number of clusters (k=5 based on elbow method)
kmeans = KMeans(n_clusters=5, random_state=42)
data_frame['Cluster'] = kmeans.fit_predict(data_frame[clustering_features])

# Calculate cluster centers and sizes
cluster_centers = kmeans.cluster_centers_
cluster_sizes = data_frame['Cluster'].value_counts().sort_index()

print("\nCluster Sizes:")
for cluster, size in cluster_sizes.items():
    print(f"Cluster {cluster}: {size} samples ({size/len(data_frame)*100:.1f}%)")

# Calculate mean values for key features in each cluster
cluster_means = data_frame.groupby('Cluster')[clustering_features].mean()
print("\nCluster Centers (Mean Values):")
print(cluster_means)

# Interpret clusters
print("\nCluster Interpretation:")
for i in range(5):
    print(f"\nCluster {i}:")
    print("Characteristics:")
    for feature in clustering_features:
        value = cluster_means.loc[i, feature]
        if abs(value) > 0.5:  # Only show significant deviations from mean
            magnitude = "high" if value > 0 else "low"
            print(f"- {magnitude.capitalize()} {feature}")
```

Visualize the clustering results using scatter plots


```python
# Create scatter plots to visualize clusters
plt.figure(figsize=(15, 10))

# Plot 1: Quantity vs UnitPrice
plt.subplot(2, 2, 1)
scatter = plt.scatter(data_frame['Quantity'], data_frame['UnitPrice'], 
                     c=data_frame['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.title('Clusters: Quantity vs UnitPrice')
plt.colorbar(scatter)

# Plot 2: Quantity vs Total Price
plt.subplot(2, 2, 2)
scatter = plt.scatter(data_frame['Quantity'], data_frame['Total Price'],
                     c=data_frame['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel('Quantity')
plt.ylabel('Total Price')
plt.title('Clusters: Quantity vs Total Price')
plt.colorbar(scatter)

# Plot 3: UnitPrice vs Total Price
plt.subplot(2, 2, 3)
scatter = plt.scatter(data_frame['UnitPrice'], data_frame['Total Price'],
                     c=data_frame['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel('UnitPrice')
plt.ylabel('Total Price')
plt.title('Clusters: UnitPrice vs Total Price')
plt.colorbar(scatter)

plt.tight_layout()
plt.show()
```

Discuss how businesses can use customer segmentation insights


```python
# Calculate and print silhouette score
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(data_frame[clustering_features], data_frame['Cluster'])
print(f"\nSilhouette Score: {silhouette_avg:.3f}")

# Interpret the score
print("\nInterpretation:")
if silhouette_avg > 0.5:
    print("Good clustering quality - clusters are well separated")
elif silhouette_avg > 0.25:
    print("Moderate clustering quality - some overlap between clusters")
else:
    print("Poor clustering quality - significant overlap between clusters")
```
